```python
import sys
import linecache

def remove_html_markup(s):
    # This is a buggy version of the function
    tag   = False
    quote = False
    out   = ""
    for c in s:
        if c == '<' and not quote:
            tag = True
        elif c == '>' and not quote:
            tag = False
        elif c == '"':
            quote = not quote
        elif not tag:
            out = out + c
    return out

def traceit(frame, event, arg):
    """This function prints out the execution trace,
    i.e. the actual contents of the code lines as executed."""
    if event == "line":
        filename = frame.f_code.co_filename
        lineno = frame.f_lineno
        print "Line {}: {}".format(lineno, linecache.getline(filename, lineno)),
    return traceit

def test():
    sys.settrace(traceit)
    remove_html_markup('"<b>foo</b>"')
    sys.settrace(None)

test()
```
